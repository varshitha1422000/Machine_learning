##**Decision Tree**
In this section Decision tree classsificaion model is built. Decision tree is one of the most popular algorithm used because of its easy understanding. I have used sklearn module in python to implement it. 



```python
# importing all Library required for this notebook

import numpy as np
import pandas as pd
from sklearn import datasets
from sklearn.tree import DecisionTreeClassifier
```

###**Loading Data**

```python
#  Reading the data

iris=datasets.load_iris()
df=pd.DataFrame(iris.data, columns=iris.feature_names)
print(df.head(5))

y=iris.target
print(y)
```

###**Training the algorithm**

```python
# Creating The model

dtree=DecisionTreeClassifier()
model=dtree.fit(df,y)

print('The Decision tree created')
```

```python
# importing the Libraries Visualize the Decision tree

from sklearn.externals.six import StringIO
from IPython.display import Image
from sklearn.tree import export_graphviz
import pydotplus
```

```python
#  Visualize graph

dot_data=StringIO()
export_graphviz(model,out_file=dot_data,feature_names=iris.feature_names,
                filled=True,rounded=True)
graph=pydotplus.graph_from_dot_data(dot_data.getvalue())
Image(graph.create_png())
```
